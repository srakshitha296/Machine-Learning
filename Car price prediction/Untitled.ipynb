```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\sraks\\AIML Internship\\Exercise\\Car price prediction\\carprices.csv')
df
```

```python
%matplotlib inline
plt.scatter(df.Mileage, df['Sell Price($)'])
plt.xlabel("Mileage")
plt.ylabel("Sell Price($)")
```

```python
dummy = pd.get_dummies(df['Car Model'])
dummy
```

```python
conc = pd.concat([df, dummy], axis='columns')
conc
```

```python
final = conc.drop(['Car Model', 'Mercedez Benz C class'], axis = "columns")
final
```

```python
x = final.drop(['Sell Price($)'], axis = "columns")
x
```

```python
y = final[['Sell Price($)']]
y
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(x,y)
```

```python
model.predict([[45000, 4, 0, 0]])
```

```python
model.predict([[86000, 7, 0, 1]])
```

```python
model.score(x,y)
```
